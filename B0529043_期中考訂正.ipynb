```python
from thinkdsp import CosSignal, SinSignal
from thinkdsp import decorate

import matplotlib.pyplot as pl
import numpy as np

cos_sig = CosSignal(freq=10, amp=10, offset=0)
sin_sig = SinSignal(freq=20,amp=10, offset=0)
```

```python
pl.grid(True)
cos_sig.plot()
decorate(xlabel='sec',)
```

```python
pl.grid(True)
sin_sig.plot()
decorate(xlabel='sec')
```

```python
mix = sin_sig + cos_sig
pl.grid(True)
mix.plot()
decorate(xlabel='sec')
```

```python
wave = mix.make_wave(duration=2,framerate=1000)
pl.grid(True)
wave.plot()
decorate(xlabel='sec')
```

```python
print(len(wave.ys))
```

```python
segment = wave.segment()
spectrum = segment.make_spectrum()

pl.grid(True)
spectrum.plot()
decorate(xlabel='Hz')
```

```python
sin1 = SinSignal(freq=1000, amp=0.25)
cos1 = CosSignal(freq=2000, amp=0.25)
sin2 = SinSignal(freq=3000, amp=0.25)
cos2 = CosSignal(freq=4000, amp=0.25)
mix = sin1 + sin2 + cos1 + cos2
mix.plot()
wave = mix.make_wave(duration=4)
spectrum = wave.make_spectrum()
pl.grid(True)
spectrum.plot()
decorate(xlabel='Hz')
```

```python
spectrum = wave.make_spectrum()
spectrum.high_pass(3000)
spectrum.plot()
pl.grid()
decorate(xlabel='Hz')
```

```python
spectrum = wave.make_spectrum()
spectrum.low_pass(2000)
spectrum.plot()
pl.grid(True)
decorate(xlabel='Hz')
```

```python
import  os 
import thinkdsp as td

a =10
samplerate = 1000
t = np.linspace(0,2,2*samplerate+1)
x = np.cos(a*t**2)
w = td.Wave(x,framerate = samplerate)
w.plot()
pl.grid(True)
```

```python
spectrogram = w.make_spectrogram(seg_length=512)
spectrogram.plot(high=20)
```
