```python
from thinkdsp import read_wave

wave = read_wave('Track5.wav')
wave.normalize()
wave.make_audio()
```

```python
wave.plot()
decorate(xlabel='Frequency (Hz)')
```

```python
wave.make_spectrogram(512).plot(high=2000)
decorate(xlabel='Time (s)', ylabel='Frequency (Hz)')
```
